#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
html = requests.get(url).content
soup = BeautifulSoup(html,'lxml')
links = soup.find_all('a')
links
````

````python
# Now let's extract only the links with href html tag
````

````python
urls = []
for link in links:
    if link.has_attr('href'):
        urls.append(link['href'])
urls                 
````

````python
# Since there are some anchor texts, let's remove them and clean up the urls
````

````python
domain = 'https://en.wikipedia.org'
rem = '#'
final_list = []

for i in urls: 
    if rem in i:
        urls.remove(i)
    elif i.startswith('https'):
        final_list.append(i)
    else:
        final_list.append(domain + i)
        
final_list     
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
html = requests.get(url).content
soup = BeautifulSoup(html,'lxml')
info = soup.find_all(['td', 'tb_region'], class_=['tabev6','tabev1','tabev2'])
info
````

````python
# What's better? Applying limit method when scraping or after having the dataframe
````

````python
info_clean = []

for i in info:
    info_clean.append(i.text.strip())

info_clean
````

````python
# Now let's get rid of all the unwanted characters
````

````python
ref = []

for i in info_clean:
    ref.append(i.replace('earthquake','').replace(u'\xa0',''))
    
ref
````

````python
clean_list = [ref[x:x+6] for x in range(0,len(ref),6)]

clean_list
````

````python
import pandas as pd
````

````python
latest_eq = []

for x in clean_list:
    f = x[0].split('.')
    date = f[0]
    time = f[1]
    latitude = x[1]+x[2]
    longitude = x[3]+x[4]
    #region = x[-1]
    latest_eq.append([date,time,latitude,longitude])

latest_eq
````

````python
df_eq = pd.DataFrame(latest_eq[:20],columns=['Date','Time','Latitude','Longitude'])
df_eq
````
